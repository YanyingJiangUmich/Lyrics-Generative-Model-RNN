```python
import json
import matplotlib.pyplot as plt
```

```python
# Load the JSON data for training and validation for Perplexity
with open('fit_20230405-182804_train.json', 'r') as json_file:
    data_train = json.load(json_file)

with open('fit_20230405-182804_validation.json', 'r') as json_file:
    data_valid = json.load(json_file)

# Extract the data from the list
timestamps = [item[0] for item in data_train]
indices = [item[1] for item in data_train]
values = [item[2] for item in data_train]

timestamps2 = [item[0] for item in data_valid]
indices2 = [item[1] for item in data_valid]
values2 = [item[2] for item in data_valid]

# Create the plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(indices, values, label='Training')
ax1.plot(indices2, values2, label='Validation')
# Customize the plot
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Perplexity')
ax1.set_title('Perplexity vs Epoch - GRU')
ax1.grid(False)
ax1.legend()

# Load the JSON data for training and validation for Loss
with open('fit_20230405-182804_train loss.json', 'r') as json_file:
    loss_train = json.load(json_file)

with open('fit_20230405-182804_validation_loss.json', 'r') as json_file:
    loss_valid = json.load(json_file)

# Extract the data from the list
timestamps_losstrain = [item[0] for item in loss_train]
indices_losstrain = [item[1] for item in loss_train]
values_losstrain = [item[2] for item in loss_train]

timestamps_lossvalid = [item[0] for item in loss_valid]
indices_lossvalid = [item[1] for item in loss_valid]
values_lossvalid = [item[2] for item in loss_valid]

# Create the plot
ax2.plot(indices_losstrain, values_losstrain, label='Training')
ax2.plot(indices_lossvalid, values_lossvalid, label='Validation')
# Customize the plot
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.set_title('Loss vs Epoch - GRU')
ax2.grid(False)
ax2.legend()

# Display the plot
plt.show()
```

```python
# Load the JSON data for training and validation for Perplexity - LSTM
with open('LSTM_per_fit_20230405-223514_train.json', 'r') as json_file:
    perplexity_train = json.load(json_file)
with open('LSTM_per_fit_20230405-223514_validation.json', 'r') as json_file:
    perplexity_valid = json.load(json_file)

with open('LSTMLOSSfit_20230405-223514_train.json', 'r') as json_file:
    loss_train = json.load(json_file)
with open('LSTMLOSSfit_20230405-223514_validation.json', 'r') as json_file:
    loss_valid = json.load(json_file)


# Extract the data from the list
timestamps_perplexity_train = [item[0] for item in perplexity_train]
indices_perplexity_train = [item[1] for item in perplexity_train]
values_perplexity_train = [item[2] for item in perplexity_train]

timestamps_perplexity_valid = [item[0] for item in perplexity_valid]
indices_perplexity_valid = [item[1] for item in perplexity_valid]
values_perplexity_valid = [item[2] for item in perplexity_valid]

timestamps_loss_train = [item[0] for item in loss_train]
indices_loss_train = [item[1] for item in loss_train]
values_loss_train = [item[2] for item in loss_train]

timestamps_loss_valid = [item[0] for item in loss_valid]
indices_loss_valid = [item[1] for item in loss_valid]
values_loss_valid = [item[2] for item in loss_valid]

# Create the plot for Perplexity
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(indices_perplexity_train, values_perplexity_train, label='Training')
ax1.plot(indices_perplexity_valid, values_perplexity_valid, label='Validation')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Perplexity')
ax1.set_title('Perplexity vs Epoch - LSTM')
ax1.grid(False)
ax1.legend()

# Create the plot for Loss
ax2.plot(indices_loss_train, values_loss_train, label='Training')
ax2.plot(indices_loss_valid, values_loss_valid, label='Validation')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.set_title('Loss vs Epoch - LSTM')
ax2.grid(False)
ax2.legend()

# Display the plot for Perplexity and Loss - LSTM
plt.show()

```
